```python
df.isna().sum()
datadrop = df.dropna()
datadrop.dropna().sum()
drop_list = ['borough_flag','code']
filtered_df = datadrop.drop(drop_list,axis=1)
filtered_df.tail()
```

```python
filtered_df.plot(kind= 'scatter',
          x='average_price',
          y ='houses_sold', 
          alpha=.1,
          c= filtered_df['no_of_crimes'], 
          cmap=plt.get_cmap('jet'), 
          colorbar=True,
          figsize=(15,10),
          sharex=False)
plt.title('Houses Sold and Average Price Against Number of Crimes')
```
